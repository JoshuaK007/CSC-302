```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

plt.figure()
x_vals = range(len(labels))
plt.bar(x_vals, men_means, width=0.3, color='blue' )
ax=plt.gca()

#adding a new bar for new data
new_xvals=[]
for item in x_vals:
  new_xvals.append(item + 0.3)

plt.bar(new_xvals, women_means, width=0.3, color='red')
ax.set_xticks(x_vals)
ax.set_xticklabels(labels)
```

```python
import numpy as np

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()

ax = plt.gca()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()



#ax.bar_label(rects1, padding=3)
#ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
x_vals = range(len(men_means))

plt.figure()
plt.barh(x_vals, men_means, height=0.3, color='b', label='Men')
plt.barh(x_vals, women_means, left=men_means, height=0.3, color='r', label='Women')

ax = plt.gca()
ax.set_yticks(x)
ax.set_yticklabels(labels)
ax.legend()
```

With matplotlib and imshow

```python
# read dataset
df = pd.read_csv('/content/drive/MyDrive/DATA//cereal.csv')
# get correlations
df_corr = df.corr()
# irrelevant fields
fields = ['rating', 'shelf', 'cups', 'weight']
# drop rows
df_corr.drop(fields, inplace=True)
# drop cols
df_corr.drop(fields, axis=1, inplace=True)

#with matplotlib

plt.figure()
plt.imshow(df_corr, cmap='hot', interpolation='nearest')
ax = plt.gca()

plt.xticks(rotation=90)
ax.set_xticks(range(len(df_corr.columns)))
ax.set_xticklabels(df_corr.columns)

ax.set_yticks(range(len(df_corr.columns)))
ax.set_yticklabels(df_corr.columns)
plt.show()
```

```python
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/mpg_ggplot2.csv")

# Draw Plot
plt.figure(figsize=(13,10), dpi= 80)
sns.distplot(df.loc[df['class'] == 'compact', "cty"], color="dodgerblue", label="Compact", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.distplot(df.loc[df['class'] == 'suv', "cty"], color="orange", label="SUV", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.distplot(df.loc[df['class'] == 'minivan', "cty"], color="g", label="minivan", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
plt.ylim(0, 0.35)

# Decoration
plt.title('Density Plot of City Mileage by Vehicle Type', fontsize=22)
plt.legend()
plt.show()
```
